```python
import numpy as np
from tensorflow.keras.models import load_model


from tensorflow.keras import layers,models
import librosa
```

```python
X_train_l = np.load('X_train_language.npy')
X_test_l = np.load('X_test_language.npy')
y_train_l = np.load('y_train_language.npy')
y_test_l = np.load('y_test_language.npy')
```

```python
model_language_4 = models.Sequential()

model_language_4.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(128, 130, 1)))
model_language_4.add(layers.BatchNormalization())
model_language_4.add(layers.MaxPooling2D((2, 2)))

model_language_4.add(layers.Conv2D(128, (3, 3), activation='relu'))
model_language_4.add(layers.BatchNormalization())
model_language_4.add(layers.MaxPooling2D((2, 2)))

model_language_4.add(layers.Conv2D(256, (3, 3), activation='relu'))
model_language_4.add(layers.BatchNormalization())
model_language_4.add(layers.MaxPooling2D((2, 2)))

model_language_4.add(layers.Flatten())

model_language_4.add(layers.Dense(128, activation='relu'))
model_language_4.add(layers.BatchNormalization())
model_language_4.add(layers.Dense(64, activation='relu'))
model_language_4.add(layers.Dense(4, activation='softmax'))

model_language_4.compile(optimizer='adam',
                       loss='categorical_crossentropy',
                       metrics=['accuracy'])
```

```python

model_language_5 = models.Sequential()

model_language_5.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(128, 130, 1)))
model_language_5.add(layers.BatchNormalization())
model_language_5.add(layers.MaxPooling2D((2, 2)))
model_language_5.add(layers.Dropout(0.25))  # Added dropout

model_language_5.add(layers.Conv2D(128, (3, 3), activation='relu'))
model_language_5.add(layers.BatchNormalization())
model_language_5.add(layers.MaxPooling2D((2, 2)))
model_language_5.add(layers.Dropout(0.25))  # Added dropout

model_language_5.add(layers.Conv2D(256, (3, 3), activation='relu'))
model_language_5.add(layers.BatchNormalization())
model_language_5.add(layers.MaxPooling2D((2, 2)))
model_language_5.add(layers.Dropout(0.25))  # Added dropout

model_language_5.add(layers.Flatten())

model_language_5.add(layers.Dense(128, activation='relu'))
model_language_5.add(layers.BatchNormalization())
model_language_5.add(layers.Dropout(0.5))  # Added dropout
model_language_5.add(layers.Dense(64, activation='relu'))
model_language_5.add(layers.Dense(4, activation='softmax'))

model_language_5.compile(optimizer='adam',
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])
```

```python

model_language_6 = models.Sequential()

model_language_6.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(128, 130, 1)))
model_language_6.add(layers.BatchNormalization())
model_language_6.add(layers.MaxPooling2D((2, 2)))

model_language_6.add(layers.Conv2D(128, (3, 3), activation='relu'))
model_language_6.add(layers.BatchNormalization())
model_language_6.add(layers.MaxPooling2D((2, 2)))

model_language_6.add(layers.Conv2D(256, (3, 3), activation='relu'))
model_language_6.add(layers.BatchNormalization())
model_language_6.add(layers.MaxPooling2D((2, 2)))

model_language_6.add(layers.GlobalAveragePooling2D())  # Replace Flatten with GlobalAveragePooling2D
model_language_6.add(layers.Dense(128, activation='relu'))
model_language_6.add(layers.BatchNormalization())
model_language_6.add(layers.Dropout(0.5))  # Added dropout

model_language_6.add(layers.Dense(64, activation='relu'))
model_language_6.add(layers.Dense(4, activation='softmax'))

model_language_6.compile(optimizer='adam',
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])
```

```python

model_language_7 = models.Sequential()

model_language_7.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(128, 130, 1)))
model_language_7.add(layers.BatchNormalization())
model_language_7.add(layers.MaxPooling2D((2, 2)))

model_language_7.add(layers.Conv2D(128, (3, 3), activation='relu'))
model_language_7.add(layers.BatchNormalization())
model_language_7.add(layers.MaxPooling2D((2, 2)))

model_language_7.add(layers.Conv2D(256, (3, 3), activation='relu'))
model_language_7.add(layers.BatchNormalization())
model_language_7.add(layers.MaxPooling2D((2, 2)))

model_language_7.add(layers.Conv2D(512, (3, 3), activation='relu'))
model_language_7.add(layers.BatchNormalization())
model_language_7.add(layers.MaxPooling2D((2, 2)))

model_language_7.add(layers.Flatten())
model_language_7.add(layers.Dense(256, activation='relu'))
model_language_7.add(layers.BatchNormalization())
model_language_7.add(layers.Dropout(0.5))

model_language_7.add(layers.Dense(128, activation='relu'))
model_language_7.add(layers.BatchNormalization())
model_language_7.add(layers.Dropout(0.5))

model_language_7.add(layers.Dense(64, activation='relu'))
model_language_7.add(layers.Dense(4, activation='softmax'))

model_language_7.compile(optimizer='adam',
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])
```

```python
ep=5
history_language_4 = model_language_4.fit(X_train_l, y_train_l, epochs=ep, validation_data=(X_test_l, y_test_l), batch_size=32)
```

```python
import matplotlib.pyplot as plt

def plot_history(histories, names):
    plt.figure(figsize=(12, 8))
    
    # Define different line styles and colors for each model
    line_styles = ['-', '--', '-.', ':']
    colors = ['blue', 'green', 'red', 'purple']

    for i, (history, name) in enumerate(zip(histories, names)):
        acc = history.history['accuracy']
        val_acc = history.history['val_accuracy']
        loss = history.history['loss']
        val_loss = history.history['val_loss']

        epochs = range(1, len(acc) + 1)

        # Plot accuracy
        plt.subplot(2, 1, 1)
        plt.plot(epochs, acc, linestyle=line_styles[i], color=colors[i], label=f'{name} Training Accuracy')
        plt.plot(epochs, val_acc, linestyle=line_styles[i], color=colors[i], marker='o', label=f'{name} Validation Accuracy')
        plt.title('Training and Validation Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()

        # Plot loss
        plt.subplot(2, 1, 2)
        plt.plot(epochs, loss, linestyle=line_styles[i], color=colors[i], label=f'{name} Training Loss')
        plt.plot(epochs, val_loss, linestyle=line_styles[i], color=colors[i], marker='o', label=f'{name} Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

    plt.tight_layout()
    plt.show()
# Assuming histories is a list containing the history objects of each language
histories = [history_language_4, history_language_5, history_language_6, history_language_7]
language_names = ['language 4', 'language 5', 'language 6', 'language 7']

plot_history(histories, language_names)
```
