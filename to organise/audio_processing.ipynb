Spectogram

```python
d=librosa.stft(y)
s_db=librosa.amplitude_to_db(np.abs(d),ref=np.max)
s_db.shape
```

```python
d=librosa.stft(x)
s_db=librosa.amplitude_to_db(np.abs(d),ref=np.max)
s_db.shape
```

```python
# Add a channel dimension
s_db_with_channel = np.expand_dims(s_db, axis=-1)
```

```python
s_db_with_channel.shape
```

```python

plt.figure(figsize=(10, 8))
plt.imshow(np.transpose(s_db_with_channel)[0], cmap='viridis', aspect='auto', origin='lower')

# Label the axes
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Spectrogram')

plt.colorbar(format='%+2.0f dB')  # Add a colorbar for reference

plt.show()
```
